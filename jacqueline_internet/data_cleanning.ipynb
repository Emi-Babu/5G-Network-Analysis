```python
# Dropping unwanted rows
internet_clean = internet.dropna()

# Renaming colums
internet_clean = internet_clean.rename(columns = { "Unnamed: 1": "Metropolitan Areas", 
                                        "Unnamed: 2" : "% Households with Internet"})
# Extracting Columns
internet_clean = internet_clean[["Metropolitan Areas","% Households with Internet"]]
# Resetting the index
internet_clean = internet_clean.reset_index(drop=True)
# Creating a new dataframe to split the Metropolitan areas
new = internet_clean["Metropolitan Areas"].str.split(", ", n = 1, expand = True) 
# Change the state names by its abrebbiation
new[1] = new[1].map(states)

# Merge the city name with the state abbreviation
internet_clean["Metropolitan Areas"] = new[0] + ", " + new[1]

# Clean Data
internet_clean
```

## Exporting data to csv file

```python
internet_clean.to_csv(r'output/internet_clean.csv')
```
