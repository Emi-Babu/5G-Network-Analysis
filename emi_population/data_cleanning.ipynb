```python
import pandas as pd 
```

## Raw Data for GDP

```python
population = pd.read_csv('source/population.csv', encoding="ISO-8859-1")
population.head()
```

## Cleanning process of Data

```python
# Dropping unwanted rows
population_clean = population

# Renaming colums
population_clean = population_clean.rename(columns = { "GC.display-label.1": "Metropolitan Areas", 
                                        "respop72018" : "Pop Est Jul 2018"})
# Extracting Columns
population_clean = population_clean[["Metropolitan Areas","Pop Est Jul 2018"]]

# Dropping unwanted rows
population_clean = population_clean.dropna()
population_clean = population_clean.drop([0,1,2])

# Resetting the index
population_clean = population_clean.reset_index(drop=True)

# Removing unwanted text
population_clean["Metropolitan Areas"] = population_clean["Metropolitan Areas"].apply(lambda x: x.replace(" Metro Area",""))


population_clean
```

## Exporting data to csv file

```python
population_clean.to_csv(r'output/population_clean.csv')
```
