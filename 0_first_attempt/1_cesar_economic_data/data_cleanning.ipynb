```python
import pandas as pd 
```

## Raw Data for GDP

```python
# Reading the csv file
gdp = pd.read_csv('source/gdp.csv') 
gdp.head()
```

## Cleanning process of Data

```python
# Dropping unwanted rows
gdp_clean = gdp.dropna()
gdp_clean = gdp_clean.drop(2)
# Renaming colums
gdp_clean = gdp_clean.rename(columns = { "Table 2. Real Gross Domestic Product (GDP) by Metropolitan Area, 2012-2017": "Metropolitan Areas", 
                                        "Unnamed: 6" : "GDP 2017"})
# Extracting Columns
gdp_clean = gdp_clean[["Metropolitan Areas","GDP 2017"]]
# Resetting the index
gdp_clean = gdp_clean.reset_index(drop=True)

gdp_clean 
```

## Exporting data to csv file

```python
gdp_clean.to_csv(r'output/gdp_clean.csv')
```

## Raw Data for GDP

```python
# Reading the csv file
unemployment = pd.read_csv('source/unemployment.csv', encoding="ISO-8859-1")
unemployment.head(10)
```

## Cleanning process of Data

```python
# Dropping unwanted rows
unemployment_clean = unemployment.dropna()
unemployment_clean = unemployment_clean.drop(0)

# Renaming colums
unemployment_clean = unemployment_clean.rename(columns = { "Unemployment Rates for Metropolitan Areas, Not Seasonally Adjusted": "Metropolitan Areas",
                                              "Unnamed: 1" : "Unemployment Rate Jul 2017"
                                              } )


# Extracting Columns
unemployment_clean = unemployment_clean[["Metropolitan Areas","Unemployment Rate Jul 2017"]]
# Resetting the index
unemployment_clean = unemployment_clean.reset_index(drop=True)
# Removing unwanted text
unemployment_clean["Metropolitan Areas"] = unemployment_clean["Metropolitan Areas"].apply(lambda x: x.replace(" Metropolitan Statistical Area",""))
unemployment_clean["Metropolitan Areas"] = unemployment_clean["Metropolitan Areas"].apply(lambda x: x.replace(" Metropolitan NECTA",""))

unemployment_clean 
```

## Exporting data to csv file

```python
unemployment_clean.to_csv(r'output/unemployment_clean.csv')
```

## Dataframes Merge

```python
merged_data = gdp_clean.merge(unemployment_clean, left_on='Metropolitan Areas', right_on='Metropolitan Areas')
merged_data
```

## Exporting data to csv file

```python
merged_data['Metropolitan Areas'] 

data_dash = merged_data[merged_data['Metropolitan Areas'].str.contains("-")]
data_dash.head(50)
```
