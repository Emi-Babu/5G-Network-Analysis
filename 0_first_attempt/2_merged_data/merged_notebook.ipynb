# Upload and Merging Process of Clean Data 
## (GDP, Unemployment, Income, Population)

### Importing Libraries

```python
# importing necesary libraries
import pandas as pd
```

## Import of Clean Data from Previous Processes

### Import of Economic Data (GDP and Unemployment)

```python
# Import of Economic Data resulting of the cleanning process
economy_df = pd.read_csv("../1_cesar_economic_data/output/economic_clean.csv")
# Drop the index loaded from the csv file
economy_df = economy_df.drop(['Unnamed: 0'], axis=1)
# Print the clean economic data
economy_df
```

### Import Income

```python
# Import of Income Data resulting of the cleanning process
wages_df = pd.read_csv("../1_pratikchhya_wages/output/wage_clean.csv")
# Drop the index loaded from the csv file
wages_df = wages_df.drop(['Unnamed: 0'], axis=1)
# Print the clean income data
wages_df
```

### Import of Internet Access Data

```python
# Import of Internet Access Data resulting of the cleanning process
internet_df = pd.read_csv("../1_jacqueline_internet/output/internet_clean.csv")
# Drop the index loaded from the csv file
internet_df = internet_df.drop(['Unnamed: 0'], axis=1)
# Print the clean Internet Acess data
internet_df
```

### Import of Population Data

```python
# Import of IPopulation Data resulting of the cleanning process
population_df = pd.read_csv("../1_emi_population/output/population_clean.csv")
# Drop the index loaded from the csv file
population_df = population_df.drop(['Unnamed: 0'], axis=1)
# Print the clean Population data
population_df
```

## Merge of Data

### Merge of Economic Data (GDP and Unemployment) and Income (wages)

```python
# Dataframes merge
total_df = pd.merge(economy_df,wages_df, on="Metropolitan Areas", how="outer")
# Resulting dataframe
total_df.head()
```

```python
# Check the data by column
total_df.count()
```

### Merge of Economic Data (GDP and Unemployment) and Income (wages) with Internet Access

```python
# Dataframes merge
total_df = pd.merge(total_df, internet_df,  on="Metropolitan Areas", how="outer")
# Resulting dataframe
total_df.head()
```

```python
# Check the data by column
total_df.count()
```

### Merge of Economic Data (GDP and Unemployment), Income (wages) anb Internet with Population

```python
# Dataframes merge
total_df = pd.merge(total_df, population_df,  on="Metropolitan Areas", how="outer")
# Resulting dataframe
total_df.head()
```

```python
# Check the data by column
total_df.count()
```

## Cleanning of Merged Data

### Cleanning

```python
# Sort data by Metropolitan Area
total_df = total_df.sort_values(by=['Metropolitan Areas'])
# Remotion of rows with empty values
total_df = total_df.dropna()
# Reset of index after dropping empty values
total_df = total_df.reset_index(drop=True)
# Print the resulting Dataframe
total_df
```

### Transforming data from Object to Float to be used to plot

#### Checking the type of data by column

```python
# Checking the type of data
total_df.dtypes
```

#### Removing "," from thousands on strings respresenting numbers

```python
# Remove commas from thousands from data
total_df['GDP 2017'] = total_df['GDP 2017'].apply(lambda x: x.replace(",",""))
total_df['Quarterly Income'] = total_df['Quarterly Income'].apply(lambda x: x.replace(",",""))
total_df['Total Income'] = total_df['Total Income'].apply(lambda x: x.replace(",",""))
total_df['Avg Income'] = total_df['Avg Income'].apply(lambda x: x.replace(",",""))
total_df['% Households with Internet'] = total_df['% Households with Internet'].apply(lambda x: x.replace(",",""))
total_df['% Households with Internet'] = total_df['% Households with Internet'].apply(lambda x: x.replace("%",""))
```

#### Transform data to Float

```python
total_df['GDP 2017'] = total_df['GDP 2017'].astype(float)
total_df['Quarterly Income'] = total_df['Quarterly Income'].astype(float)
total_df['Total Income'] = total_df['Total Income'].astype(float)
total_df['Avg Income'] = total_df['Avg Income'].astype(float)
total_df['% Households with Internet'] = total_df['% Households with Internet'].astype(float)
```

#### Verify they type of data

```python
# Verify the type of data
total_df.dtypes
```

#### Adding the GDP per Capita from values in two Columns

```python
# New column for GDP per Capita
total_df["GDP Per Capita"] =  total_df["GDP 2017"] / total_df["Pop Est Jul 2018"]
# Print of the resulting data frame
total_df
```

## Export Data

```python
# Export of Final Data Merged to be used in Plotting
total_df.to_csv(r'output/clean_data.csv')
```
