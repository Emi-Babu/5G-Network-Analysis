```python
import pandas as pd
```

```python
economy_df=pd.read_csv("../cesar_economic_data/output/economic_clean.csv")
economy_df = economy_df.drop(['Unnamed: 0'], axis=1)
economy_df.head()
```

```python
wages_df = pd.read_csv("../pratikchhya_wages/output/wage_clean.csv")
wages_df = wages_df.drop(['Unnamed: 0'], axis=1)
wages_df = wages_df.reset_index(drop=True)
wages_df.head()
```

```python
internet_df = pd.read_csv("../jacqueline_internet/output/internet_clean.csv")
internet_df = internet_df.drop(['Unnamed: 0'], axis=1)
internet_df = internet_df.reset_index(drop=True)
internet_df.head()
```

```python
total_df = pd.merge(economy_df,wages_df, on="Metropolitan Areas", how="outer")
total_df.head()
```

```python
total_df = pd.merge(total_df, internet_df,  on="Metropolitan Areas", how="outer")
total_df.head()
```

```python
total_df.count()
```

```python
total_df = total_df.dropna()
total_df.head()
```

```python
total_df.count()
```

```python
total_df = total_df.reset_index(drop=True)
total_df.head()
```

```python
total_df.to_csv(r'output/clean_data.csv')
```
