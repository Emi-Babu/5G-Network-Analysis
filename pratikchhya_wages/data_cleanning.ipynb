```python
import pandas as pd 
```

```python
gdp = pd.read_csv('source/wages.csv') 
gdp.head()
```

```python
wages_df=gdp[["Metropolitan", "Quarterly","Total","Average"]]
wages_df.head(10)
```

```python
wages_df.count()
```

```python
wage=wages_df.iloc[6:]
wage.head()
```

```python
wage.dropna()
wage.count()
```

```python
wage = wage.reset_index(drop=True)
wage.head()
```

```python
wage_pd= wage.rename({"Average":"Avg Income", "Quarterly":"Quarterly Income","Total":"Total Income", "Metropolitan":"Metropolitan Areas"}, axis=1)
wage_pd.head()
```

```python
# Removing unwanted text
wage_pd["Metropolitan Areas"] = wage_pd["Metropolitan Areas"].apply(lambda x: x.replace(" MSA",""))
wage_pd.head()
```

```python
# Export data
wage_pd.to_csv(r'output/wage_clean.csv')
```
