```python
economy_data1=pd.read_csv("cesar_economic_data/output/economic_clean.csv")
economy_data1.head()
```

```python
gdp_data2=pd.read_csv("cesar_economic_data/output/gdp_clean.csv")
gdp_data2.head()
```

```python
unemployment_data3=pd.read_csv("cesar_economic_data/output/unemployment_clean.csv")
unemployment_data3.head()
```

```python
wages_file= pd.read_csv("pratikchhya_wages/output/wage_clean.csv")
wages_file.head()
```

```python
economy_gdp=pd.merge(economy_data1,gdp_data2, on="Metropolitan Areas", how="outer")
economy_gdp.head()
```

```python
economy_gdp.count()
```

```python
unemploy_ecogdp=pd.merge(economy_gdp,unemployment_data3, on="Metropolitan Areas", how="outer")
unemploy_ecogdp.head()
```

```python
unemploy_ecogdp.count()
```

```python
wage_all= pd.merge(unemploy_ecogdp,wages_file,on="Metropolitan Areas", how="outer")
wage_all.head()
```

```python
wage_all.count()
```

```python
clean_file= wage_all.dropna()
clean_file.head()
```
